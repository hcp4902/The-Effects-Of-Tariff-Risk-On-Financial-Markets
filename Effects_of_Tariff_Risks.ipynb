```python
import yfinance as yf
import pandas as pd
import numpy as np
```

**Load Data**

```python
tariff_days = pd.read_excel('/content/Data (Jan25-Mar25).xlsx',sheet_name= 'Tariff Days')
tariff_days.set_index('Date',inplace=True)
# tariff_days[['10 year Breakeven Inflation (pp)','2 year Treasury yield (pp)','10 year Treasury yield (pp)']] = tariff_days[['10 year Breakeven Inflation (pp)','2 year Treasury yield (pp)','10 year Treasury yield (pp)']] *100
tariff_days.head()
```

```python
pre_tariff_days = pd.read_excel('/content/Data (Jan25-Mar25).xlsx',sheet_name= 'Pre-announcement Days')
pre_tariff_days.set_index('Date',inplace=True)
# pre_tariff_days[['10 year Breakeven Inflation (pp)','2 year Treasury yield (pp)','10 year Treasury yield (pp)']] = pre_tariff_days[['10 year Breakeven Inflation (pp)','2 year Treasury yield (pp)','10 year Treasury yield (pp)']] *100

pre_tariff_days.head()
```

**Analysis**

```python
tariff_days_covariance = tariff_days.T.dot(tariff_days)#tariff_days.cov()
tariff_days_covariance
```

```python
pre_tariff_announcemnt_covariance = pre_tariff_days.T.dot(pre_tariff_days)#pre_tariff_days.cov()
pre_tariff_announcemnt_covariance
```

```python
delta_cov = tariff_days_covariance - pre_tariff_announcemnt_covariance
delta_cov
```

**Mean values**

```python
# Assuming you have the mean values
tariff_mean = tariff_days.mean()
pre_tariff_mean = pre_tariff_days.mean()
means = pd.concat([tariff_mean,pre_tariff_mean],axis=1)
means
```

```python
# Assuming you have the mean values
tariff_var = tariff_days.var()
pre_tariff_var = pre_tariff_days.var()
variance = pd.concat([tariff_var,pre_tariff_var, (tariff_var/pre_tariff_var-1)*100],axis=1)
variance
```

**IV Estimators**

```python
d = pd.DataFrame(index = tariff_days.columns,columns = ['Estimator 1', 'Estimator 2'])
x1 = 'High Yield Spread'

for security in tariff_days.columns:
  # Ensure x1 and security are valid keys before accessing
  d.loc[security, 'Estimator 1'] = 0.6*delta_cov.loc[security,security]/delta_cov.loc[x1,security]
  d.loc[security, 'Estimator 2'] = 0.6*delta_cov.loc[x1,security]/delta_cov.loc[x1,x1]
  d.loc[security, 'cov Explained'] = 100*((delta_cov.loc[x1,security]/delta_cov.loc[x1,x1])**2)*delta_cov.loc[x1,x1]/tariff_days_covariance[security][security]
  # d.loc[security, 'cov'] = delta_cov.loc[security,security]

d
```
